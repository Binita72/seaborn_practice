#seaborn

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Read csv file
diamonds=pd.read_csv(r"C:\Users\Mou\Desktop\finity\diamonds.csv")
```

```python
diamonds
```

```python
sns.set_style("darkgrid", {'axes.grid' : True})
sns.lmplot(x= 'carat', y='depth', data=diamonds)
plt.show()
```

```python
sns.lmplot(x='carat', y='depth', data=diamonds, hue='color',palette='Set2')
plt.show()
```

```python
sns.lmplot(x='carat', y='depth', data=diamonds, col='color')
plt.show()
```

Univariate â†’ â€œone variableâ€ data visualization

```python
#strip plot

sns.stripplot(y= 'depth', data=diamonds,jitter=False)
plt.ylabel('depth')
plt.show()
```

```python
diamonds.head()
```

```python
#Grouping with stripplot()
sns.stripplot(x='price', y='carat', data=diamonds)
plt.ylabel('price')
plt.show()
```
